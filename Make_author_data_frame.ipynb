```python
start_year = 1980
actual_year = 2010
df_clean_NEW = df_clean[(df_clean['year'] >= start_year) & (df_clean['year'] <= actual_year)]

NO_sample = 500000 #1080316 #numer of sample rows taken for analysis

#comment out to take random rows
rows = random.sample(df_clean_NEW.index, NO_sample)
df_clean_AUX = df_clean_NEW.ix[rows]

# or comment out to take last rows as sample
#df_clean_AUX = df_clean_NEW.tail(NO_sample)

print df_clean.shape
print df_clean_NEW.shape
print df_clean_AUX.shape

# usuwanie mem
df_clean = 0
df_clean_NEW = 0
```

```python
start = time()
# memoize all authors and their papers

authors_by_year = {}

for row in df_clean_AUX.iterrows():
    row_num, actual_row = row
    auth_list = actual_row['tot_auth_list']
    actual_year = actual_row['year']
    actual_paper = str(actual_row['recid'])
    
    if type(auth_list) is list:
        auth_list = [author.lower() for author in auth_list]  #convert to lowercase to collect only truly unique names    
        for author in auth_list:
            author_year = (author, actual_year)

            #search if author_year has already been seen          
            if not author_year in authors_by_year :
                authors_by_year[author_year] = {}
                authors_by_year[author_year]['Author'] = author
                authors_by_year[author_year]['year'] = actual_year
                authors_by_year[author_year]['NO_papers'] = 0
                authors_by_year[author_year]['recid_papers'] = ''
                authors_by_year[author_year]['collaborators'] = ()
                authors_by_year[author_year]['NO_collaborators'] = 0
            authors_by_year[author_year]['NO_papers'] += 1
            authors_by_year[author_year]['recid_papers'] = authors_by_year[author_year]['recid_papers'] +'::'+ actual_paper
            authors_by_year[author_year]['collaborators'] = authors_by_year[author_year]['collaborators'] + tuple(auth_list)
            authors_by_year[author_year]['NO_collaborators'] += 1

end = time()
print end - start
```
